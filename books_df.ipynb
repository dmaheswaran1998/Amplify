```python
book_df=book_df.dropna()
book_df.count()
category_list=book_df["Categories"].to_list()
book_list=book_df["Title"].to_list()
```

```python
book_list1=[]
region_list=[]
for x in range(0,len(category_list)):
    item = str(category_list[x])
    split=item.split(",")
    for i in range(0, len(split)):
    
        str1= split[i]
        str2 = '/region'

        if str2 in str1:
                region_list.append(str1)
                book_list1.append(book_list[x])
                
    
```

```python
book_list1
```

```python
countries_list=[]

for i in range (0, len(region_list)):
    txt=region_list[i].split("/")
    countries_list.append(txt[-1])
```

```python
countries_list
countries_list_edited=[]
for item in countries_list:
    item=item.replace("-"," ")
    countries_list_edited.append(item)
```

```python
countries_list_edited
```

```python
print(len(countries_list_edited))
```

```python
book_df_edited=pd.DataFrame({"Title":book_list1, "Country":countries_list_edited})
```

```python
book_df_edited
```

```python
book_df_edited["Lat"]=""
book_df_edited["Lng"]=""
```

```python
book_df_edited
```

```python
for index, row in book_df_edited.iterrows():
    
    search_string=row["Country"]
    
    try:
        url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={search_string}&inputtype=textquery&fields=photos,formatted_address,name,rating,opening_hours,geometry&key={gkey}'
        # Run request
        response = requests.get(url)
        search_geo = response.json()
        
    
        book_df_edited.loc[index, 'Lat'] = search_geo["candidates"][0]['geometry']['location']['lat']
        book_df_edited.loc[index, "Lng"]=search_geo["candidates"][0]['geometry']['location']['lng']

    
    #in the case of a JSONDecode Error, this ensures the code will keep running
    except (KeyError, IndexError, ValueError):
        book_df_edited.loc[index, 'Lat'] = "No Coordinates"
        book_df_edited.loc[index, 'Lng'] = "No Coordinates"
```

```python
book_df_edited.to_csv("book_geocoordinates.csv")
```

```python
#creating the map json dictionary 

title_list=book_df_edited["Title"].to_list()
lat=book_df_edited["Lat"].to_list()
lng=book_df_edited["Lng"].to_list()
```

```python
final_list=[]
for i in range(0,len(title_list)):
    coordinates=[lng[i], lat[i]]
    first_dict={"Title":title_list[i], "Latitude":lat[i], "Longitude":lng[i], "Coordinates":coordinates}
    final_list.append(first_dict)
    
```

```python
#write this out out as a json 
with open('book_map.json', 'w') as fp:
    json.dump(final_list, fp)
```
